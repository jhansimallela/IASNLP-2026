## 2. Basic Data Types, Boolean Logic, and Truthy/Falsy Values
Python has standard data types: `int`, `float`, `str`, and `bool`.

**Truthy vs. Falsy:** In Python, you don't always need an explicit `== True` check.
- **Falsy values:** `0`, `None`, empty strings `""`, and empty collections `[]`, `{}`, `()`.
- **Truthy values:** Almost everything else.

```python
is_active = True
empty_list = []

# Boolean logic uses 'and', 'or', 'not'
if is_active and not empty_list:
    print("This runs because is_active is True AND empty_list is Falsy (so 'not Falsy' is True).")

# Truthy/Falsy in action
if "Hello":
    print("Non-empty strings evaluate to True!")
```

## 3. Collections: Lists, Tuples, Dictionaries, and Sets
Python provides four main built-in data structures to store collections of data.

```python
# 1. LIST (Mutable, Ordered)
tech_stack = ["C++", "Python", "JavaScript"]
tech_stack.append("React")
print(f"List: {tech_stack}")

# 2. TUPLE (Immutable, Ordered) - Cannot be changed after creation
coordinates = (17.4401, 78.3489)
print(f"Tuple: {coordinates}")

# 3. SET (Unordered, Unique elements) - Great for removing duplicates rapidly
unique_ids = {101, 102, 101, 103, 102}
print(f"Set (Duplicates removed): {unique_ids}")

# 4. DICTIONARY (Key-Value pairs, Mutable)
heroes_team = {
    "team_name": "Avengers",
    "members": ["Iron Man", "Captain America", "Thor"],
    "enemy": "Thanos"
}
print(f"Team {heroes_team['team_name']} is fighting {heroes_team['enemy']}.")
```

## 4. Built-in Functions: len, range, enumerate, zip
**Note:** `zip` combines iterables and stops at the shortest one, while `enumerate` provides an automatic counter/index so you don't have to manage it manually.

```python
names = ["Superman", "Batman", "Wonder Woman"]
abilities = ["Strength", "Intellect", "Agility"]

print(f"Total team members (len): {len(names)}")

print("\n--- Using range ---")
for i in range(3): # Generates numbers 0, 1, 2
    print(f"Index {i}")

print("\n--- Using enumerate ---")
for index, name in enumerate(names):
    print(f"Hero {index + 1}: {name}")

print("\n--- Using zip ---")
for name, ability in zip(names, abilities):
    print(f"{name} has the ability: {ability}")
```

## 5. Control Flow: if/else, for, and while loops
Controlling the execution path of your program.

```python
# If / Elif / Else
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Grade: C")

# While loop with a break condition
print("\nStarting while loop:")
count = 0
while count < 5:
    if count == 3:
        print("Count is 3, breaking out of the loop!")
        break
    print(f"Current count: {count}")
    count += 1
```

## 6. List Comprehensions & Generator Expressions
Pythonic ways to create iterables in a single line of code.

```python
# List comprehension: [expression for item in iterable if condition]
squares = [x**2 for x in range(10) if x % 2 == 0]
print(f"Even squares (List): {squares}")

# Generator expression: uses () instead of [].
# Evaluates lazily! Highly memory efficient for massive datasets.
squares_gen = (x**2 for x in range(10) if x % 2 == 0)
print(f"Generator object: {squares_gen}")
print(f"First yielded value: {next(squares_gen)}")
print(f"Second yielded value: {next(squares_gen)}")
```

## 7. Functions, Exception Handling & Debugging
**Why use try/except?** It prevents the entire script from crashing when it hits an error, which is crucial for writing robust, production-ready code.

```python
def safe_divide(a, b):
    """Divides two numbers with basic error checking."""
    try:
        result = a / b
        return result
    except ZeroDivisionError as e:
        # Basic debugging/error checking print
        print(f"[Debug] Cannot divide by zero. Error details: {e}")
        return None
    except TypeError:
        print("[Debug] Inputs must be numerical values.")
        return None

print("Success case:", safe_divide(10, 2))
print("Failure case:", safe_divide(10, 0))
```

## 8. Basic Object-Oriented Programming (Classes)
Classes bundle data (attributes) and behavior (methods) together to model real-world entities.

```python
class AlgorithmSolver:
    # The __init__ method initializes the object's state
    def __init__(self, language):
        self.language = language
        self.problems_solved = 0

    # A class method
    def solve(self):
        self.problems_solved += 1
        print(f"Solved an algorithmic problem in {self.language}!")

# Creating an instance (object) of the class
coder = AlgorithmSolver("C++")
coder.solve()
coder.solve()
print(f"Total solved: {coder.problems_solved}")
```

## 9. File Handling, JSON, Modules & Imports
Interacting with the file system and leveraging external/built-in modules.

```python
import json # Built-in module for JSON parsing
import os   # Built-in module for OS-level operations

config_data = {"model": "Gemma", "type": "LLM", "parameters": "7B"}

# 1. File Handling & JSON Writing using context managers (the 'with' statement)
with open("model_config.json", "w") as file:
    json.dump(config_data, file, indent=4)

# 2. Reading JSON back
with open("model_config.json", "r") as file:
    loaded_data = json.load(file)

print(f"Successfully loaded config for {loaded_data['model']}")
```

```python
#Clean up the demo file
if os.path.exists("model_config.json"):
    os.remove("model_config.json")
```

## 10. Data Science Ecosystem: NumPy, Pandas, and Matplotlib
**Heads Up:** If you are running this locally, you may need to install these packages in your environment using `pip install numpy pandas matplotlib`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# NumPy: Fast, vectorized array operations (C-level performance)
arr = np.array([1, 2, 3, 4, 5])
print(f"NumPy array multiplied by 10:\n{arr * 10}\n")

# Pandas: Data manipulation and analysis (DataFrames)
df = pd.DataFrame({
    "Dish": ["Butter Chicken", "Dal Makhani", "Chole Bhature", "Naan"],
    "Rating": [9.5, 8.8, 9.2, 9.0]
})
print("Pandas DataFrame (Delhi Food Ratings):")
print(df)

# Matplotlib: Basic Data Visualization
# Uncomment the lines below to render the plot!

plt.figure(figsize=(8, 4))
plt.bar(df["Dish"], df["Rating"], color='orange')
plt.title("Food Ratings")
plt.xlabel("Dish")
plt.ylabel("Rating (/10)")
plt.ylim(0, 10)
plt.show()
```
